```python
import sys
import time

import numpy as np
import cv2 as cv
```

```python
def main():
    cap = cv.VideoCapture(0)
    
    cv.namedWindow("Input", cv.WINDOW_AUTOSIZE)
    cv.namedWindow("Output", cv.WINDOW_AUTOSIZE)
    
    while True:
        _, srcFrame = cap.read()
        # convert bgr frame to gray frame
        grayFrame = cv.cvtColor(srcFrame, cv.COLOR_BGR2GRAY)
        
        cv.imshow("Input",srcFrame)
        cv.imshow("Output",grayFrame)
        
        if cv.waitKey(1) & 0xFF == ord('q'):
            print(srcFrame.shape[0]) # 480 x 600
            
            # Accessing pixel intensity values
            w, h, _ = srcFrame.shape
            print(srcFrame[w//2, h//2, 0])
            
            break
            
    cap.release()
    cv.destroyAllWindows()
        
        
```

```python
if __name__ == "__main__":
    main()
```
